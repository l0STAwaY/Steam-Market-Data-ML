```python

from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.chrome.options import Options
import time
import random


    
    
def get_Y_scroll_position(driver):
    return driver.execute_script("return window.pageYOffset")



def scroll_bottom(driver):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
    time.sleep(random.uniform(1, 5))

    

# Recommended or not/hours played/actual review/ early acess or not
def get_player_data(card, game_id, name):
    # Get the player data from the individual card
    date = get_date_pc(card)
    player_count = get_player_count_pc(card)
    gain = get_gain_pc(card)
    peak_player = get_peakplayer(card)

    # Construct the player data in a dictionary (or any other structure you prefer)
    player_data = {
        "game_id": game_id,
        "name": name,
        "date": date,
        "player_count": player_count,
        "gain": gain,
        "peak_player":peak_player
     
    }
    
    return player_data
    

# https://stackoverflow.com/questions/47660523/finding-elements-by-class-name-with-spaces

def get_date_pc(card):
    try:
       return card.find_element(By.CSS_SELECTOR, ".month-cell.left").text
   
    except NoSuchElementException:
        return "Not Found"  # Return empty string if element is not found

def get_player_count_pc(card):
    try:
        return card.find_element(By.CSS_SELECTOR, ".right.num-f").text        

    except NoSuchElementException:
        return card.find_element(By.CSS_SELECTOR, ".right.num-p.gainorloss.gain").text
        return "Not Found"  # Return empty string if element is not found
    
def get_gain_pc(card):
    try:

        return card.find_element(By.CSS_SELECTOR, ".right.num-p.gainorloss.loss").text
    except NoSuchElementException:
        # If "loss" element is not found, try the "gain" element
        try:
            
            return card.find_element(By.CSS_SELECTOR, ".right.num-p.gainorloss.gain").text
        except NoSuchElementException:
            # If neither element is found, print a message
            print("Neither loss nor gain element found")
            
            
            
            


def get_peakplayer(card):
    try:
       return card.find_element(By.CSS_SELECTOR, ".right.num").text
   
    except NoSuchElementException:
        return "Not Found"  # Return empty string if element is not found

    

# card = driver.find_elements(By.CLASS_NAME, "odd")[1]

# print(card.text)

# date = get_date_pc(card)
# player_count = get_player_count_pc(card)
# gain = get_gain_pc(card)
# peak_player = get_peakplayer(card)

# # Print the results outside the functions
# print("Date:", date)
# print("Player Count:", player_count)
# print("Gain:", gain)
# print("Peak Player:", peak_player)

    

# element = driver.find_elements(By.XPATH, "//tbody[@class='tabular-nums']/tr")
# card = element[0]

# print(get_date_pc(card))
# print(get_player_count_pc(card))
# print(get_gain_pc(card))
# print(get_percentage_gain_pc(card))

```

```python


def scrap_player_data(driver,game_id,name):
    try:
        previous_position = get_Y_scroll_position(driver)
        state = True
        scroll_count = 0
        while(state):
            element = driver.find_elements(By.CLASS_NAME, "odd")
       
            print(f"Found {len(element)} cards")

            
            for card in element:
                review_package = get_player_data(card, game_id,name)
                reviews.append(review_package)
                
            while(scroll_count < max_scroll):
                scroll_bottom(driver)
                current_position = get_Y_scroll_position(driver) 
                scroll_count+=1
                  
                if scroll_count % 200 == 0:
                    print(f"Scroll count: {scroll_count}")    
                # print(f"Scroll count: {scroll_count}")          

                if scroll_count >= max_scroll:
                    print("Reached max scroll limit. Stopping.")
                    state = False
                    break
                
                # This means we reached the end of the scrolling page
                elif previous_position == current_position:
                        print("No movement detected, possibly end of page.")
                        break  # Break from inner loop to re-check elements
                    # safer random break less likly to get blocked
                    
                    # time.sleep(random.uniform(1, 2))
                    
                    # give the program a bit slack maybe its the loading time issue
                     
                else:
                    previous_position = current_position
                    break
                
                
    except Exception as e:
        # Handles the exception
        print(f"An error occurred: {e}")
```
