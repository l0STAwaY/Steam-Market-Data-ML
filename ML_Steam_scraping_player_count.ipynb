```python

from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.chrome.options import Options
import time
import random



def get_date_pc(card):
    try:
       return card.find_element(By.XPATH, ".//td[1]").text
   
    except NoSuchElementException:
        return "Not Found"  # Return empty string if element is not found

def get_player_count_pc(card):
    try:
    
        return card.find_element(By.XPATH, ".//td[2]").text        

    except NoSuchElementException:
        return "Not Found"  # Return empty string if element is not found
    
def get_gain_pc(card):
    try:
       return card.find_element(By.XPATH, ".//td[3]").text
   
    except NoSuchElementException:
        return "Not Found"  # Return empty string if element is not found

    
def get_percentage_gain_pc(card):
    try:
       return card.find_element(By.XPATH, ".//td[4]").text
   
    except NoSuchElementException:
        return "Not Found"  # Return empty string if element is not found
    
    
    

element = driver.find_elements(By.XPATH, "//tbody[@class='tabular-nums']/tr")
card = element[0]

print(get_date_pc(card))
print(get_player_count_pc(card))
print(get_gain_pc(card))
print(get_percentage_gain_pc(card))



# get macro data


def get_followers(app_chart_numbers):
    try:
        return app_chart_numbers.find_elements(By.TAG_NAME, "li")[0].text
    except IndexError:
        return "Not Found"

def get_rank(app_chart_numbers):
    try:
        return app_chart_numbers.find_elements(By.TAG_NAME, "li")[1].text
    except IndexError:
        return "Not Found"

def get_positive_reviews(app_chart_numbers):
    try:
        return app_chart_numbers.find_elements(By.TAG_NAME, "li")[2].text
    except IndexError:
        return "Not Found"

def get_negative_reviews(app_chart_numbers):
    try:
        return app_chart_numbers.find_elements(By.TAG_NAME, "li")[3].text
    except IndexError:
        return "Not Found"

def get_positive_percentage(app_chart_numbers):
    try:
        return app_chart_numbers.find_elements(By.TAG_NAME, "li")[4].text
    except IndexError:
        return "Not Found"

def extract_app_chart_info():
    app_chart_numbers = driver.find_element(By.CLASS_NAME, "app-chart-numbers")
    
    followers = get_followers(app_chart_numbers)
    rank = get_rank(app_chart_numbers)
    positive_reviews = get_positive_reviews(app_chart_numbers)
    negative_reviews = get_negative_reviews(app_chart_numbers)
    positive_percentage = get_positive_percentage(app_chart_numbers)

    print(followers)
    print(rank)
    print(positive_reviews)
    print(negative_reviews)
    print(positive_percentage)


extract_app_chart_info()
```
