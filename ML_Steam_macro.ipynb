```python

from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.chrome.options import Options
import time
import random
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

    
    
def get_Y_scroll_position(driver):
    return driver.execute_script("return window.pageYOffset")



def scroll_bottom(driver):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
    time.sleep(random.uniform(1, 5))

    

# Recommended or not/hours played/actual review/ early acess or not
def get_player_data(card, game_id, name):
    # Get the player data from the individual card
    
    # This is very annoying, took me a while to realize that we have to use java script to make the hidden attribute visible to scrape
    value = driver.execute_script("""
    var elems = document.querySelectorAll('span.nonresponsive_hidden.responsive_reviewdesc');
    if (elems.length > 1) {
        elems[1].style.display = 'block';  // Make the second element visible by changing 'none' to 'block'
    }
    return elems.length > 1 ? elems[1].style.display : 'Element not found';
    """)
   
    elements = WebDriverWait(driver, 10).until(
    EC.presence_of_all_elements_located((By.CSS_SELECTOR, "span.nonresponsive_hidden.responsive_reviewdesc"))
    )[1].text
    
    
    positive_percent = elements.split()[1]
    total_review = elements.split()[4]
    

    # Construct the player data in a dictionary (or any other structure you prefer)
    player_data = {
        "game_id": game_id,
        "name": name,
        "total_review": total_review,
        "positive_percent": positive_percent,    
    }
    
    return player_data
    

# https://stackoverflow.com/questions/47660523/finding-elements-by-class-name-with-spaces



def get_positive_percent_pc(card):
    try:
        return driver.find_element(By.CSS_SELECTOR, ".nonresponsive_hidden.responsive_reviewdesc").text      

    except NoSuchElementException:
        return "Not Found"  # Return empty string if element is not found





```

```python


def scrap_player_data(driver,game_id,name):
    try:
        previous_position = get_Y_scroll_position(driver)
        state = True
        scroll_count = 0
        while(state):

        

            review_package = get_player_data(driver, game_id,name)
            reviews.append(review_package)
                
            while(scroll_count < max_scroll):
                scroll_bottom(driver)
                current_position = get_Y_scroll_position(driver) 
                scroll_count+=1
                  
                if scroll_count % 200 == 0:
                    print(f"Scroll count: {scroll_count}")    
                # print(f"Scroll count: {scroll_count}")          

                if scroll_count >= max_scroll:
                    print("Reached max scroll limit. Stopping.")
                    state = False
                    break
                
                # This means we reached the end of the scrolling page
                elif previous_position == current_position:
                        print("No movement detected, possibly end of page.")
                        break  # Break from inner loop to re-check elements
                    # safer random break less likly to get blocked
                    
                    # time.sleep(random.uniform(1, 2))
                    
                    # give the program a bit slack maybe its the loading time issue
                     
                else:
                    previous_position = current_position
                    break
                
                
    except Exception as e:
        # Handles the exception
        print(f"An error occurred: {e}")
```
